```python
#importing the libaries..

import time
import datetime
from datetime import date

import os
import random
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


```

```python
from dateutil.relativedelta import relativedelta
from pandas.tseries.offsets import DateOffset
from pandas.tseries.offsets import MonthEnd
```

```python
# importing the NSE libaries (nsepy) financial data

from nsepy import get_history
from nsepy.derivatives import get_expiry_date
```

```python
# We are trying to import BANK NIFTY index future by using nsepy library between 1 march 2022 to 31 march 2022...

banknifty_future = get_history(symbol='BANKNIFTY',
                        start=date(2022,3,1),
                        end=date(2022,3,30),
                        index=True,
                        futures=True,
                        expiry_date=date(2022,3,31))
banknifty_future
```

```python
#changing the date index into a string..

banknifty_future.reset_index(inplace=True)
banknifty_future['Date'] = banknifty_future['Date'].astype(str)
```

```python
# PLOTTING THE CLOSING PRICE OF BANK NIFTY FUTURE INDEX (1 march to 31 march 2022)

plt.figure(figsize=(12,6))
plt.style.use('ggplot')
plt.plot(banknifty_future.Date,banknifty_future.Close,"o-b")
plt.xticks(rotation=90)
plt.title("Closing price of BANK NIFTY Futures")
plt.ylabel("Price")
plt.xlabel('Date')
plt.show()
```

```python
# PLOTTING THE VOLUME OF BANK NIFTY FUTURE INDEX (1 march to 31 march 2022)

plt.figure(figsize=(12,6))
plt.bar(banknifty_future.Date,banknifty_future.Volume, edgecolor="black", color= "blue")
plt.xticks(rotation=90)
plt.title("Volume of BANK NIFTY Futures")
plt.ylabel("Volume")
plt.xlabel('Date')
plt.show()
```
